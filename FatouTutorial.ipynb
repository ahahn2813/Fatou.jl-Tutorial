# The $\texttt{juliafill()}$ Method

This method generates the Fatou, Julia, and Filled in Julia sets for a function $f(z).$ Remember this means we are looking at the behavior of the iterates $f^m(z)$ in neighborhoods about $z.$ We must then use $\texttt{plot}$ from the $\texttt{PyPlot}$ package to produce the graphic.

This method has $14$ parameters, but those important to use are:

      ∂::Array{Float64,1}                 # Bounds, $[x(a),x(b),y(a),y(b)]$
      n::Integer  = 176,                  # horizontal grid points
      N::Integer  = 35,                   # max. iterations
      iter::Bool  = false,                # toggle whether to color the image by iteration count
      p::Number   = 0,                    # iteration color exponent
      cmap::String= "",                   # imshow color map

For cmap color choices see: https://matplotlib.org/stable/users/explain/colors/colormaps.html

#### Source Code Investigation

Creation of $\texttt{Fatou.Define}$ objects is done via passing a function expression (in variables z, c) and optional keyword arguments to the $\texttt{juliafill(), mandelbrot()}$, and $\texttt{newton()}.$

# The $\texttt{mandelbrot()}$ Method

The famous Mandelbrot set is the numbers $c\in \mathbb{C}$ for which the function $f_c(z) = z^2 + c$ does not diverge to infinity when iterated and evaluated at $0.$ In other words, it is the set of all $c\in \mathbb{C}$ such that the sequence of numbers $(f_c^n(0))_{n=0}^\infty$ remains bounded for all $n>0.$

### History

This set was first defined and drawn by Robert W. Brooks and Peter Matelski in 1978, as part of a study of Kleinian groups. The first image to be generated of the Mandelbrot set was by Benoit Mandelbrot in $1980$ while working at IBM's Watson Research Center in New York. The region of * is the Mandelbrot set.

```julia
mandFrac = mandelbrot(:(z^2+c),n=800,N=20,∂=[-1.91,0.51,-1.21,1.21],iter = true, cmap="gist_earth")
plot(fatou(mandFrac))
```

# The $\texttt{newton()}$ Method

Recall that Newton's method is used approximate the roots of a function. If we begin at $x_0,$ then we can find the equation of the line tangent to $f(x)$ at $x_0:$
$$y-f(x_0) = f'(x_0)(x-x_0).$$
As we want to find the root, we see where the line intersects the $x$-axis, or rather we set $y = 0$ and rearranging, we know
$$x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}.$$
We can hopefully get closer to a root, if we repeat the proccess by finding the line tangent to $f(x)$ at the point $x_1.$ Doing this we find we are simply iterating, getting the recursion process
$$x_{m+1} = x_m−\frac{f(x_m)}{f'(x_m)},$$



Under certain assumptions, this process will converge to the root, but there are situations where
we run into trouble. So, it would be helpful to know for what ``starting points" do we converge to a root, and when will we not.

In the complex setting with a holomorphic map $f,$ we can view this as the iteration of the map which takes $z\mapsto z−\frac{f(z)}{f'(z)}.$ Thus, we look at this family of iterates and construct the corresponding Fatou ande Julia sets. This is called the $\textbf{Newton Fractal}$ for $f.$ In other words, we look at the Fatou and Julia set for the iterates of $g(z) = z-\frac{f(z)}{f'(z)}.$

The $\texttt{newton()}$ method allows us to easily see for what ``starting points" we converge and where we converge to.

### Example 1: $f(z) = z^3-1$

We can easily find that the roots are the third roots of unity: $1,-\frac{1}{2}+i\frac{\sqrt{3}}{2},$ and $-\frac{1}{2}-i\frac{\sqrt{3}}{2}.$

Each root lies in the largest colored regions. For example, the root $1$ lies in the large green region on the right. Moreover, points which will converge to that root are contained in that large green region. Same is true for the large red and blue regions for the remaining two roots.

Something interesting is occuring on the boundary. The smaller green regions in the boundary contain points which will converge to the root the root $1$. Likewise for the smaller red and blue regions.

```julia
nf = newton(:(z^3-1),n=800,ϵ=.01,N=100,cmap="turbo")
plot(fatou(nf))
```

### Example 2: $f(z) = z^3-2z-5$

# Orbits

### Orbit Definitions

Let $G\subset \mathbb{C}$ and $f:G\rightarrow G$ a function. The $\textbf{forward orbit}$ of $x_0\in G$ under $f$ is the set
    \begin{align*}
        O^+_f(x_0)& = \{f^m(x_0):m\in \mathbb{N}\}.
    \end{align*}
The $\textbf{backward orbit}$ of $x_0\in G$ under $f$ is defined as the set
    \begin{align*}
        O^-_f(x_0)& = \{w\in \mathbb{C}\cup \{\infty\}: f^m(w) = x_0, m\in \mathbb{N}\}.
    \end{align*}

### Example 1: Orbit of $\phi(z) = z^2-0.67$ at $z_0 = 1.25$

Thus, the iterates are
\begin{align*}
    \phi^1(z) & = z^2-0.67\\
    \phi^2(z) & = (z^2-0.67)^2-0.67\\
    \phi^3(z) & = ((z^2-0.67)^2-0.67)^2-0.67\\
    \vdots &
\end{align*}

Rough calculation would be:
\begin{align*}
    O_f^+(1.25) = \{1.25,.8925,.12655,-.6539,\ldots\}
\end{align*}

```julia
orbit(juliafill(:(z^2-0.67),∂=[-1.25,1.5],x0=1.25,orbit=100,depth=3,n=147))
```

The grey "x's" map the set of points in the forward orbit $O_f^+(1.25).$
